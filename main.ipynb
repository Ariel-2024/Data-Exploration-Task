```python
import pandas as pd


netflix_file_path = "./netflix_titles.csv"

netflix_data = pd.read_csv(netflix_file_path)

#Ratio of Movies to TV Shows
movie_ratio = (netflix_data['type'] == 'Movie').sum()
tv_show_ratio = (netflix_data['type'] == 'TV Show').sum()

#print("Ratio of Movies to TV Shows:",movie_ratio,":", tv_show_ratio)
print("Ratio of Movies to TV Shows;",movie_ratio/tv_show_ratio,": 1 \n")


#How it has changed over the years(Ratio over the years) Calculate the ratio: TV Shows / Movies
print("The Ratio of Movies To TV show has increased over the years")
netflix_data["release_year"].value_counts()
counts = netflix_data.groupby(['release_year', 'type']).size().unstack(fill_value=0)
counts['tv_to_movie_ratio'] = counts['TV Show'] / counts['Movie']
print(counts, "\n")

#Top Ten countries by number of production(TV Shows and Movies)
print("Top Ten Countries by number of production:")
print(netflix_data['country'].value_counts().head(10),"\n")

#Top ten countries by movie
movie_counts = netflix_data[netflix_data['type'] == 'Movie']['country'].value_counts().head(10)

#Top ten countries by TV Show
tv_counts = netflix_data[netflix_data['type'] == 'TV Show']['country'].value_counts().head(10)
#print("Top Ten Countries by number of production of movies \n", movie_counts)
#print("Top Ten Countries by number of production of TV Shows\n",tv_counts)



#Month of the year with more release compared to other months
netflix_data['month']= netflix_data['date_added'].str.split(' ').str[0]
number_max = netflix_data['month'].value_counts().max()
max_month = netflix_data['month'].value_counts().idxmax()
print("Yes, The month July has the highest number of releases on Netflix compared to other months.")
print(max_month, "", number_max, "\n")

#Genre for both movies and TV shows
netflix_data['split_listed_in'] = netflix_data['listed_in'].str.replace('&', ',').str.split(',')
netflix_data_exploded = netflix_data.explode('split_listed_in')
netflix_data_exploded['split_listed_in'] = netflix_data_exploded['split_listed_in'].str.strip()
#print(netflix_data_exploded['split_listed_in'].value_counts())

#Splitted data for Genre of Movie and Tv Shows
print("Some genres appear only in movies and not in TV shows, while others are more commonly found in TV shows than in movies, and vice versa.")
listed_in_split = netflix_data_exploded.groupby(['split_listed_in', 'type']).size().unstack(fill_value=0)
print(listed_in_split,"\n")


#NO 5 Question And Answer
print("An Interesting fact about this data is that the United States produces the most netflix content for both movies and TV Show \nAlso Movies are more produced and released more than " \
"TV shows because the platform has approximately 2.29 movies for every 1 TV show.")
```
